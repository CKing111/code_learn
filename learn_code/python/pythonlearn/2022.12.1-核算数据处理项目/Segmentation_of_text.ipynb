```python
import openpyxl
import pandas as pd

xlsx_path = '.\output2.xlsx'

# # 读取单个文件数据
# def load_Data(data_path,sheet=0):   
#     df=pd.ExcelFile(data_path)        # 读取excel文件数据
#     data_sheet = df.sheet_names
#     return df,data_sheet

# df,num = load_Data(xlsx_path)
# print(num)
# df1 = df.parse(num[-1])
# print(df1)

# data = pd.DataFrame(res.reshape(1,-1))
# print(data)


import pandas as pd
import os
from openpyxl import Workbook
from openpyxl import load_workbook
# newdata = pd.DataFrame(res)#需要新写入的数据
# df = pd.read_excel(xlsx_path,sheet_name=0) #读取原数据文件和表
# writer = pd.ExcelWriter(xlsx_path,engine='openpyxl')
# if os.path.exists(xlsx_path):
#     book = load_workbook(xlsx_path)
# else:
#     book = Workbook()
# # book = load_workbook(xlsx_path)
# writer.book = book # openpyxl.workbook.workbook.Workbook
# writer.sheets = dict((ws.title, ws) for ws in book.worksheets) #{'st4': <Worksheet "st4">, 'st5': <Worksheet "st5">, 'st3': <Worksheet "st3">}
# df_rows = df.shape[0] #获取原数据的行数
# newdata.to_excel(writer, sheet_name=0,startrow = df_rows + 1, index = False, header = False)#将数据写入excel中的st3,从第一个空行开始写
# writer.save()#保存



def write_excel_xlsx(path, sheet_name, value):
    index = len(value)
    workbook = openpyxl.Workbook()  # 新建工作簿（默认有一个sheet？）
    sheet = workbook.active  # 获得当前活跃的工作页，默认为第一个工作页
    # sheet.title = sheet_name  # 给sheet页的title赋值
    for i in range(0, index):
        for j in range(0, len(value[i])):
            sheet.cell(row=i + 1, column=j + 1, value=str(value[i][j]))  # 行，列，值 这里是从1开始计数的
    workbook.save(path)  # 一定要保存
    workbook.close()
    print("xlsx格式表格写入数据成功！")
# write_excel_xlsx(xlsx_path,0,res.reshape(1,-1))

def Additional_write_excel_xlsx(path, sheet_name, res):
    xlsx_path = path
    data = openpyxl.load_workbook(xlsx_path)

    # 取第一张表
    sheetnames = data.get_sheet_names()
    table = data.get_sheet_by_name(sheetnames[0])

    table = data.active
    print(table.title)  # 输出表名
    nrows = table.max_row  # 获得行数
    ncolumns = table.max_column  # 获得列数
    values = res.reshape(1,-1)
    # 注意行业列下标是从1开始的
    for i in range(1, len(values)+1):
        for j in range(1, len(values[i-1])+1):
            table.cell(nrows+i, j).value = values[i-1][j-1]

    data.save(xlsx_path)
    data.close()

```

```python
import re  
pattern = r"([。？！… ； 】 \\n]+)" #正则匹配模式,用+表示至少一个字符
flags = ["。","？","！","…","……","；",":","】"]
blank = ["\n"]
sentence_txt = []
with open("./test1.txt","r",encoding="GBK") as reader_file:
    for line in reader_file:#一行就是一篇文章
        print('1111',line)
        if line == blank:
            segment = ""
        else:
            spilt_list = re.split(pattern=pattern, string=line) # 多字符切割
            print('2222',spilt_list)
            segment = ""
            for segment_i in spilt_list:   # 读取单个字符 
                # print('3333',segment_i)
                segment += segment_i
                if segment_i in flags : # 判定是否需要空格
                # if segment_i[-1] in flags or len(segment_i)>1:# or segment_i[-2:-1] in flags:
                    print('4444',segment_i[-1])
                    #去除分割子句中的空格,\n,\t等符号,并加上"\r"回车符换行
                    if segment.split()!= blank:
                        sentence_txt.append("".join(segment.split())+"\r") 
                        segment = ""
        # sentence_txt.append("\r")
        print('~~~~~~~~~~~~~~~~~~~~~~~~~~~~')

print(sentence_txt)
```
