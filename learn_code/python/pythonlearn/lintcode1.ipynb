```python
class Solution:
    def highestFrequency(self,ipLines):
        ans = ipLines[0]
        mapdic = {}
        for ip in ipLines:
            if ip in mapdic:
                mapdic[ip] += 1
            else:
                mapdic[ip] = 1
            max = 0
            
            for key in mapdic:
                if int(mapdic[key])>max:
                    max = mapdic[key]
                    ans = key
                return print(ans)
            
ipLines = input('请输入：')
print（ipLines，highestFrequency(ipLines)）
# ipLines  = ["192.168.1.1","192.118.2.1","192.168.1.1"]
# print(ans)
```
