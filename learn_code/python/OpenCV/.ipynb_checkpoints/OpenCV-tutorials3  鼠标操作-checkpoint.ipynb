## 双击画圈 

```python

import cv2
import numpy as np

# mouse callback function 鼠标响应及回调
def draw_circle(event,x,y,flags,param):
    #操作变量，指针触发坐标，拖拽事件，
    if event == cv2.EVENT_LBUTTONDBLCLK: #左键双击
        cv2.circle(img,(x,y),100,(255,0,0),-1)  #做圆

# Create a black image, a window and bind the function to window
img = np.zeros((512,512,3), np.uint8)
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)
        # 鼠标事件呼叫函数，需要一个自定义handler
        #窗口名字，定义的鼠标相应函数以及回调函数，
while(1):
    cv2.imshow('image',img)
    if cv2.waitKey(20) & 0xFF == 27:  #ESC退出
        break
cv2.destroyAllWindows()
```

## 鼠标绘图  

```python
import cv2
import numpy as np

drawing = False  # 当鼠标按下变为True
mode = True # 按下‘m’变成False，true为绘制矩形，false为画笔
ix,iy = -1,-1 #初始化鼠标位置

# mouse callback function
def draw_circle(event,x,y,flags,param):
    global ix,iy,drawing,mode  #全局变量，可被所有对象调用

    #左键摁下后确定初始位置
    if event == cv2.EVENT_LBUTTONDOWN:   
        drawing = True  
        ix,iy = x,y
    
    #鼠标左键摁下移动绘图
#     elif event == cv2.EVENT_MOUSEMOVE:
#         if drawing == True:  #左键摁下
#             if mode == True:  #
#                 cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1) #绘制矩形
#             else:
#                 cv2.circle(img,(x,y),5,(0,0,255),-1) #绘制圆圈
    #摁住左键拖拽画矩形
    elif event == cv2.EVENT_MOUSEMOVE:
    #and flags == cv2.EVENT_FLAG_LBUTTON:  # 当按下左键拖拽鼠标时
        if drawing is True:
            if mode is True:
                cv2.rectangle(img, (ix, iy), (x, y), (0, 255, 0), 1)  # 当模式为True时画矩形
            else:
                cv2.circle(img, (x, y), 3, (0, 0, 255), -1)  # 当模式为False时画线
    #滚动滑轮画实心圆
    elif  event == cv2.EVENT_MOUSEWHEEL:
            if drawing is True:
                if mode is True:
                    cv2.circle(img, (x, y), 50, (255, 0, 0), -1)  # 滚轮画实心圆
    #双击画空心圆
    elif event == cv2.EVENT_LBUTTONDBLCLK:
            if drawing is True:
                if mode is True:
                    cv2.circle(img, (x, y), 50, (255, 255, 0), 3)  # 双击左键画空心圆

                
    #左键谈起停止绘画
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        #下代码代表原地点击时会做小像素圆点
        if mode == True:
            cv2.rectangle(img,(ix,iy),(x,y),(255,0,0),5)
        else:
            cv2.circle(img,(x,y),5,(0,0,255),-1)
            
# Create a black image, a window and bind the function to window
img = np.zeros((512,512,3), np.uint8)
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)  #鼠标操作与窗口绑定

# while(1):
#     cv2.imshow('image',img)
#     if cv2.waitKey(20) & 0xFF == 27:
#         break
# cv2.destroyAllWindows()

while(1):
    cv2.imshow('image',img)
    k=cv2.waitKey(20) & 0xFF
    #将m键与模式转换绑定在一起
    if k ==ord('m'):
        mode=not mode
    elif k==ord('q'):
        break
cv2.destroyAllWindows()
```

##  滑动条做调色板

```python
import cv2
import numpy as np
def nothing(x):
    pass

#创建一个黑色图像
img = np.zeros((300,512,3),np.uint8)
cv2.namedWindow('image')
# 三个滑动栏
cv2.createTrackbar('R','image',0,255,nothing)
cv2.createTrackbar('G','image',0,255,nothing)
cv2.createTrackbar('B','image',0,255,nothing)
        #滑动栏名字，窗口名字，范围，输入
#用滑动条实现开关功能
switch = '0:OFF\n1:ON'
cv2.createTrackbar(switch,'image',0,1,nothing)

while(1):
    cv2.imshow('image',img)  #绑定窗口和空白图片
    k=cv2.waitKey(1) 
    if k == ord('q'):#按q键退出
        break
#绑定滑动栏和窗口（窗口已于空白图片绑定）
    r = cv2.getTrackbarPos('R','image') 
    g = cv2.getTrackbarPos('G', 'image')
    b = cv2.getTrackbarPos('B', 'image')
    s = cv2.getTrackbarPos(switch, 'image')
        #滑动栏名字，窗口名字
    if s == 0:
        img[:]=0
    else:
        img[:]=[r,g,b]
cv2.destroyAllWindows()
```
